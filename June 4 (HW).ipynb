```python
import nltk
from nltk import word_tokenize
```

```python
#importing from website
from urllib import request 
url = "http://www.gutenberg.org/files/62303/62303-0.txt"
response = request.urlopen(url) #response is received from the given url
rawtext = response.read().decode('utf8') #This function decodes a string, previously encoded with the utf8_encode() function, back to ISO-8859-1.
type(rawtext) #shows the type of text
```

```python
len(rawtext) #length of the text
```

```python
rawtext[:75]
```

```python
#tokenize the string (raw text)
tokens = word_tokenize(rawtext)
type(tokens)
```

```python
len(tokens) #length of the tokens which is being reduced fron the length of the raw text
```

```python
tokens[:20]
```

```python
text = nltk.Text(tokens) #here the nltk text is created
type(text)
```

```python
text[1024:1062] #some middle text which is tokenized
```

```python
text.concordance('odor') #finding some concordance for word odor
```

```python
from nltk.tag import DefaultTagger #loadind libraries
tagging = DefaultTagger('NN') #defining the tag for noun
tagging.tag(text)
```

```python
#Install Text Blob Module which helps for part of Speech tagging
#pip install -U textblob
#python -m textblob.download_corpora

from textblob import TextBlob
blob_object = TextBlob(rawtext) #raw text must be given not tokens
print(blob_object.tags)
```

```python
#stemming
from nltk.stem import PorterStemmer
porter = PorterStemmer()
stem_sentence = []
def stemsentence(text):
    for word in tokens:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
x = stemsentence(text)
print(x)
```
