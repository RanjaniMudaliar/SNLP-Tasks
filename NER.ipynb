```python
#importing from website
import requests 
url = "http://www.gutenberg.org/files/62303/62303-0.txt"
response = requests.get(url) #response is received from the given url
```

```python
type(response)
```

```python
html_page = response.text
```

```python
#for cleaning the text
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_page, 'html.parser')
type(soup)
```

```python
#cleaned doc 
doc = soup.get_text()
print(doc)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
#Remove the spaces and punctuations
nlp = nlp(doc[:99999])   #first 99999 words out of 11lakh words
for token in nlp:
    if not token.is_punct | token.is_space:
        print(token.text)    
```

```python
#Pos-tagging
for token in nlp:
    tag_token = (token, token.tag_)
    print(tag_token)
```

```python
#find the NER
for token in nlp.ents:
    print(token, token.label_)
```

```python
#find most common Entities from the NER
from collections import Counter
E = []
for entity in nlp.ents:
    E.append(entity.label_)
E_count = Counter(E)
top = E_count.most_common(10)
print(top)
```
